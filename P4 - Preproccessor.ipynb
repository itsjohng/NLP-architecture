```python
import PyPDF2
import textract
import re
import string
import csv
import numpy as np
import pandas as pd
```

```python
def pdf2text(filename):
    # filename is pdf document read from disk.
    # output single string of converted text. Its ugly.
    pdfFileObj = open(filename,'rb')
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj)

    num_pages = pdfReader.numPages
    count = 0
    text = ""

    while count < num_pages:
        pageObj = pdfReader.getPage(count)
        count +=1
        text += pageObj.extractText()
    # if PyPDF2 cannot read scanned files ... 
    if text != "":
        text = text
        return text
    # ... we run the OCR library textract to convert scanned/image based PDF files into text.
    else:
        text = textract.process(filename, method='tesseract', language='eng')
        text = text.decode()
        return text

def preprocessor(content):
    # simple pre-processor. The variance in document type here really limits what is possible.
    # removes punctuation, carriage returns, weird ascii characters generated by pdf converter.
    # output is a single string to be CV'd or otherwise.
    content = content.replace('\n', '')
    content = re.sub('[%s]'% re.escape(string.punctuation), ' ', content.lower())
    content = content.encode("ascii", "ignore")
    content = content.decode()
    
    return content
```

```python
doc_count = range(1,115)
doc_master = []
for doc in doc_count:
    filenam = 'Downloads/Disciplinarity/doc{}.pdf'.format(doc)
    print(filenam, " processing...")
    content = preprocessor(pdf2text(filenam))
    print('success!')
    doc_master.append(content)
    with open('docdump.csv', mode='w') as docdump:
        writer = csv.writer(docdump, quotechar='"', quoting=csv.QUOTE_MINIMAL)
        writer.writerow([doc_master])
    print(doc_master[(doc-1)][:40])
```
